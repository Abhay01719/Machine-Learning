```python
def cost(x,y,m):
    temp = 0
    for i in range(len(m)):
        temp += m[i]*x[:,i]
    return ((y-temp)**2).mean()
```

```python
def derivative(x,y,m,i):
    temp = 0
    for j in range(len(m)):
        temp += m[j]*x[j]
    return -2*((y-temp)*x[i])
    
```

```python
def gradient(x,y,m,alpha):
    temp = [0]*(len(m))
    for i in range(len(m)):
        m[i] -= alpha*derivative(x,y,m,i)
    return m
```

```python
def run(x,y):
    m = [0]*x.shape[1]
    alpha = 0.00000001
    iterations = 10
    for i in range(iterations):
        for j in range(x.shape[0]):
            m = gradient(x[j],y[j],m,alpha)
            print(cost(x,y,m))
    return m
    
```

```python
run(feature, label)
```
