```python
feature = np.array(df.HouseAge)
```

```python
label = np.array(data.target)
```

```python
def derivative(x,y,m,c):
    m_out = -2*(((y-m*x-c)*x).mean())
    c_out = -2*((y-m*x-c).mean())
    return m_out, c_out
```

```python
def cost(x,y,m,c):
    return ((y-(m*x + c))**2).mean()
        
```

```python
def run(x,y):
    m = -1
    c = 1
    l = 0.001
    for i in range(100):
        m_cal,c_cal = derivative(x,y,m,c)
        m -= l*m_cal
        c -=  l*c_cal
        print(cost(x,y,m,c))
#         print(m,c)
    return m,c
```

```python
run(feature,label)
```
